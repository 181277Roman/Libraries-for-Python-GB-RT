Задание 2:

```python
t = np.linspace(0, 10, 51)
print(t)
f = np.cos(t)

legend_font = {'size': 12,
              'family': 'calibri',
              }
plt.title('Графикf(t)', fontdict=legend_font)
plt.xlabel('Значения "t"', fontdict=legend_font)
plt.ylabel('Значения "f"', fontdict=legend_font)

plt.axis([-0.5, 9.5, -2.5, 2.5])

plt.plot(t, f, color='green')
plt.show()
```

Задание 3:

```python
x = np.linspace(-3, 3, 51)
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)

ax1.plot(x, y1)
ax1.set_title("График (x**2)")

ax2.plot(x, y2)
ax2.set_title("График (2*x + 0.5)")

ax3.plot(x, y3)
ax3.set_title("График (-3*x - 1.5)")

ax4.plot(x, y4)
ax4.set_title("График (sin(x))")
```
